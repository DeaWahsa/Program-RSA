```python
import numpy as np

def softmax(z):
    # Menghitung softmax
    exp_scores = np.exp(z)
    return exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

def compute_gradient(y_pred, y_true):
    # Menghitung gradien loss menggunakan cross-entropy loss
    n_samples = y_pred.shape[0]
    grad = softmax(y_pred)
    grad[range(n_samples), y_true] -= 1
    grad /= n_samples
    return grad
```

```python
class DecisionNode:
    def __init__(self, feature_idx=None, threshold=None, value=None, left=None, right=None):
        self.feature_idx = feature_idx
        self.threshold = threshold
        self.value = value
        self.left = left
        self.right = right

class DecisionTree:
    def __init__(self, max_depth=1):
        self.max_depth = max_depth

    def compute_gini(self, y):
        # Menghitung gini impurity
        classes = np.unique(y)
        n_samples = len(y)
        gini = 1.0
        for cls in classes:
            p = np.sum(y == cls) / n_samples
            gini -= p ** 2
        return gini

    def split(self, X, y, feature_idx, threshold):
        # Memisahkan dataset berdasarkan feature dan threshold
        left_mask = X[:, feature_idx] < threshold
        right_mask = ~left_mask
        left_X, left_y = X[left_mask], y[left_mask]
        right_X, right_y = X[right_mask], y[right_mask]
        return left_X, left_y, right_X, right_y

    def find_best_split(self, X, y):
        # Mencari pemisahan terbaik berdasarkan gini impurity
        n_features = X.shape[1]
        best_gini = float('inf')
        best_feature_idx, best_threshold = None, None
        for feature_idx in range(n_features):
            thresholds = np.unique(X[:, feature_idx])
            for threshold in thresholds:
                left_X, left_y, right_X, right_y = self.split(X, y, feature_idx, threshold)
                gini = (len(left_y) * self.compute_gini(left_y) + len(right_y) * self.compute_gini(right_y)) / len(y)
                if gini < best_gini:
                    best_gini = gini
                    best_feature_idx = feature_idx
                    best_threshold = threshold
        return best_feature_idx, best_threshold

    def build_tree(self, X, y, depth=0):
        # Rekursif membangun decision tree
        if depth >= self.max_depth or len(np.unique(y)) == 1:
            leaf_value = np.mean(y, axis=0)
            return DecisionNode(value=leaf_value)

        feature_idx, threshold = self.find_best_split(X, y)
        left_X, left_y, right_X, right_y = self.split(X, y, feature_idx, threshold)
        left = self.build_tree(left_X, left_y, depth + 1)
        right = self.build_tree(right_X, right_y, depth + 1)
        return DecisionNode(feature_idx=feature_idx, threshold=threshold, left=left, right=right)

    def predict(self, x, node):
        # Memprediksi kelas menggunakan decision tree
        if node.value is not None:
            return node.value
        if x[node.feature_idx] < node.threshold:
            return self.predict(x, node.left)
        else:
            return self.predict(x, node.right)
```

```python
class XGBoostClassifier:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.estimators = []

    def fit(self, X, y):
        # Fitting XGBoost dengan menggunakan weak learner berbasis decision tree
        y_pred = np.zeros_like(y, dtype=float)

        for _ in range(self.n_estimators):
            gradient = compute_gradient(y_pred, y)
            tree = DecisionTree(max_depth=self.max_depth)
            tree_root = tree.build_tree(X, gradient)
            self.estimators.append(tree_root)

            update = self.learning_rate * tree.predict(X, tree_root)
            y_pred += update

    def predict_proba(self, X):
        # Memprediksi probabilitas kelas menggunakan ensemble model
        y_pred = np.zeros((X.shape[0], len(self.estimators)))

        for i, estimator in enumerate(self.estimators):
            y_pred[:, i] = self.learning_rate * estimator.predict(X, estimator)

        return softmax(y_pred)

    def predict(self, X):
        # Memprediksi kelas menggunakan ensemble model
        proba_xgb = self.predict_proba(X)
        return np.argmax(proba_xgb, axis=1)
```

```python
import numpy as np

def softmax(z):
    # Menghitung softmax
    exp_scores = np.exp(z)
    return exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

def compute_gradient(y_pred, y_true):
    # Menghitung gradien loss menggunakan cross-entropy loss
    n_samples = y_pred.shape[0]
    grad = softmax(y_pred)
    grad[range(n_samples), y_true] -= 1
    grad /= n_samples
    return grad
```

```python
class DecisionNode:
    def __init__(self, feature_idx=None, threshold=None, value=None, left=None, right=None):
        self.feature_idx = feature_idx
        self.threshold = threshold
        self.value = value
        self.left = left
        self.right = right

class DecisionTree:
    def __init__(self, max_depth=1):
        self.max_depth = max_depth

    def compute_gini(self, y):
        # Menghitung gini impurity
        classes = np.unique(y)
        n_samples = len(y)
        gini = 1.0
        for cls in classes:
            p = np.sum(y == cls) / n_samples
            gini -= p ** 2
        return gini

    def split(self, X, y, feature_idx, threshold):
        # Memisahkan dataset berdasarkan feature dan threshold
        left_mask = X[:, feature_idx] < threshold
        right_mask = ~left_mask
        left_X, left_y = X[left_mask], y[left_mask]
        right_X, right_y = X[right_mask], y[right_mask]
        return left_X, left_y, right_X, right_y

    def find_best_split(self, X, y):
        # Mencari pemisahan terbaik berdasarkan gini impurity
        n_features = X.shape[1]
        best_gini = float('inf')
        best_feature_idx, best_threshold = None, None
        for feature_idx in range(n_features):
            thresholds = np.unique(X[:, feature_idx])
            for threshold in thresholds:
                left_X, left_y, right_X, right_y = self.split(X, y, feature_idx, threshold)
                gini = (len(left_y) * self.compute_gini(left_y) + len(right_y) * self.compute_gini(right_y)) / len(y)
                if gini < best_gini:
                    best_gini = gini
                    best_feature_idx = feature_idx
                    best_threshold = threshold
        return best_feature_idx, best_threshold

    def build_tree(self, X, y, depth=0):
        # Rekursif membangun decision tree
        if depth >= self.max_depth or len(np.unique(y)) == 1:
            leaf_value = np.mean(y, axis=0)
            return DecisionNode(value=leaf_value)

        feature_idx, threshold = self.find_best_split(X, y)
        left_X, left_y, right_X, right_y = self.split(X, y, feature_idx, threshold)
        left = self.build_tree(left_X, left_y, depth + 1)
        right = self.build_tree(right_X, right_y, depth + 1)
        return DecisionNode(feature_idx=feature_idx, threshold=threshold, left=left, right=right)

    def predict(self, x, node):
        # Memprediksi kelas menggunakan decision tree
        if node.value is not None:
            return node.value
        if x[node.feature_idx] < node.threshold:
            return self.predict(x, node.left)
        else:
            return self.predict(x, node.right)
```

```python
class LightGBMClassifier:
    def __init__(self, n_estimators=100, learning_rate=0.1, max_depth=1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.estimators = []

    def fit(self, X, y):
        # Fitting LightGBM dengan menggunakan weak learner berbasis decision tree
        y_pred = np.zeros_like(y, dtype=float)

        for _ in range(self.n_estimators):
            gradient = compute_gradient(y_pred, y)
            tree = DecisionTree(max_depth=self.max_depth)
            tree_root = tree.build_tree(X, gradient)
            self.estimators.append(tree_root)

            update = self.learning_rate * tree.predict(X, tree_root)
            y_pred += update

        self.y_pred = y_pred

    def predict_proba(self, X):
        # Memprediksi probabilitas kelas menggunakan ensemble model
        y_pred = np.zeros((X.shape[0], len(self.estimators)))

        for i, estimator in enumerate(self.estimators):
            y_pred[:, i] = estimator.predict(X, estimator)

        return softmax(self.learning_rate * y_pred)

    def predict(self, X):
        # Memprediksi kelas menggunakan ensemble model
        proba_lgb = self.predict_proba(X)
        return np.argmax(proba_lgb, axis=1)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import GradientBoostingClassifier

# Inisialisasi model-model yang akan digunakan
xgb = XGBClassifier(max_depth=3, n_estimators=100, reg_alpha=0.1, reg_lambda=0.1)
lgbm = LGBMClassifier(max_depth=3, n_estimators=100, reg_alpha=0.1, reg_lambda=0.1)
gb = GradientBoostingClassifier(max_depth=3, n_estimators=100)

# Latih model-model dengan metode validasi silang
xgb_scores = cross_val_score(xgb, X_train, y_train, cv=5)
lgbm_scores = cross_val_score(lgbm, X_train, y_train, cv=5)
gb_scores = cross_val_score(gb, X_train, y_train, cv=5)

# Hitung rata-rata skor validasi silang
xgb_mean_score = np.mean(xgb_scores)
lgbm_mean_score = np.mean(lgbm_scores)
gb_mean_score = np.mean(gb_scores)

# Cetak skor validasi silang untuk masing-masing model
print("Skor validasi silang XGBoost:", xgb_mean_score)
print("Skor validasi silang LightGBM:", lgbm_mean_score)
print("Skor validasi silang Gradient Boosting:", gb_mean_score)

# Fungsi objektif untuk optimisasi Firefly
def objective_function(weights):
    # Normalisasi bobot
    weights = np.exp(weights) / np.sum(np.exp(weights))

    # Latih model-model dengan data pelatihan
    xgb.fit(X_train, y_train)
    lgbm.fit(X_train, y_train)
    gb.fit(X_train, y_train)

    # Melakukan prediksi probabilitas menggunakan masing-masing model dan bobot
    ensemble = (
        weights[0] * xgb.predict_proba(X_test)
        + weights[1] * lgbm.predict_proba(X_test)
        + weights[2] * gb.predict_proba(X_test)
    )

    # Menghitung akurasi menggunakan bobot
    y_pred = np.argmax(ensemble, axis=1)
    accuracy = accuracy_score(y_test, y_pred)

    # Mengembalikan nilai negatif akurasi untuk optimisasi maksimum
    return -accuracy

# Fungsi untuk menghitung jarak antara dua firefly
def distance(firefly1, firefly2):
    return np.linalg.norm(firefly1 - firefly2)

# Fungsi untuk memperbarui posisi firefly
def update_firefly(current_firefly, best_firefly, alpha, beta, gamma):
    distance_factor = distance(current_firefly, best_firefly)
    random_movement = alpha * (np.random.random_sample() - 0.5)
    attraction_factor = beta * np.exp(-gamma * distance_factor ** 2)

    updated_firefly = (
        current_firefly
        + attraction_factor * (best_firefly - current_firefly)
        + random_movement
    )
    return updated_firefly

# Inisialisasi populasi firefly
population_size = 50
num_features = 3
population = np.random.rand(population_size, num_features)

# Melakukan iterasi dan pembaruan firefly
max_iterations = 100
best_fitness = float("-inf")
best_weights = None

for _ in range(max_iterations):
    # Evaluasi kualitas setiap firefly
    fitness = np.zeros(population_size)
    for i, firefly in enumerate(population):
        fitness[i] = objective_function(firefly)

    # Mencari firefly dengan kualitas terbaik
    max_index = np.argmax(fitness)
    best_firefly = population[max_index]
    best_fitness = fitness[max_index]

    # Pembaruan firefly
    for i in range(population_size):
        for j in range(population_size):
            if fitness[i] < fitness[j]:
                population[i] = update_firefly(
                    population[i], population[j], alpha=0.5, beta=1, gamma=1
                )

    # Memperbarui bobot terbaik jika ditemukan firefly dengan kualitas yang lebih baik
    if best_weights is None or best_fitness < max(fitness):
        best_weights = best_firefly

# Normalisasi bobot terbaik jika tidak kosong
if best_weights is not None:
    best_weights = np.exp(best_weights) / np.sum(np.exp(best_weights))
else:
    print("Tidak ada firefly yang memperbarui bobot terbaik")

# Menampilkan bobot terbaik dan akurasi model dengan bobot tersebut
print("Bobot terbaik:", best_weights)

# Implementasi bobot ke model ensemble jika tidak kosong
if best_weights is not None:
    ensemble = (
        best_weights[0] * xgb.predict_proba(X_test)
        + best_weights[1] * lgbm.predict_proba(X_test)
        + best_weights[2] * gb.predict_proba(X_test)
    )
    y_pred = np.argmax(ensemble, axis=1)
    accuracy = accuracy_score(y_test, y_pred)
    print("Akurasi Ensemble dengan bobot terbaik:", accuracy)
else:
    print("Tidak ada firefly yang memperbarui bobot terbaik")
```
